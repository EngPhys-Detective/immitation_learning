```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
```

```python
data_dir = '/home/fizzer/enph_ws/src/immitation_learning/data/'
dirt_dir = data_dir + 'dirt_track/'
paved_dir = data_dir + 'paved_track/'
```

```python
# plot 20 random images from the dirt track in a 4x5 grid
fig = plt.figure(figsize=(20, 10))
for i, file in zip(range(0,20), os.listdir(dirt_dir + '/forward/')):
    img = cv2.imread(dirt_dir + '/forward/' + file)
    fig.add_subplot(4, 5, i+1)
    # convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # select the colour white
    lower = np.array([140, 140, 140])
    upper = np.array([205, 205, 205])
    # threshold the image
    mask = cv2.inRange(img, lower, upper)
    # apply the mask
    gray = cv2.bitwise_and(gray, gray, mask=mask)

    plt.imshow(gray, cmap='gray')
plt.show()
```
