```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
path = '/home/fizzer/enph_ws/src/immitation_learning/data'
target = '/home/fizzer/enph_ws/src/immitation_learning/processed_data' 
```

```python
img = cv2.imread(path + '/forward/1_1022.png')
print(img.shape)
plt.imshow(img[220:720,:,:])
print(type(img) == np.ndarray)
```

```python
for dir in os.listdir(path):
    for file in os.listdir(path + '/' + dir):
        img = cv2.imread(path + '/' + dir + '/' + file)
        if file.find('_') == -1:
            cv2.imwrite(os.path.join(target,dir,'2_' + file), img[220:720,:,:])
        else:
            cv2.imwrite(os.path.join(target,dir,file), img[220:720,:,:])
```
